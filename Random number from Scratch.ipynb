# Generating random numbers from scratch

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_seed():
    """
     seed: previously value number generated by the generator
     output: returned by the current system clock as the seed
    """
    
    import time
    t = time.perf_counter() # current system clock
    return int(10**9*float(str(t-int(t))[0:]))
```

```python
def pseudo_uniform(mult=16807, 
                   mod=(2**31)-1, 
                   seed=123456789, 
                   size=1):
    """
    a pseudo random genarator with multipliers and modulus
    mult: Lehmer random number generator parameter (7**5)
    """
    u = np.zeros(size)
    x = (seed*mult+1)%mod
    u[0] = x/mod
    
    for i in range(1,size):
        x = (x*mult+1)%mod
        u[i] = x/mod
    return u
```

```python
ls = pseudo_uniform(size=10000, seed=3)
plt.hist(ls)
plt.show()
```

```python
def uniform(low=0, high=1, seed=123456789, size=1):
    """
    return uniformly generated random number between low and high boundaries
    """
    
    return low + (high-low)*pseudo_uniform(seed=seed, size=size)
```

```python
ls = uniform(low=-5, high=8, size=10000)
plt.hist(ls, bins=20, edgecolor='k')
plt.show()
```

```python
for i in range(5):
    print(uniform(low=-5, high=7, size=10, seed=generate_seed()))
```

## Visualization using statistical methods

### Binomial distribution

```python
def binomial_distribution(n=100, p=0.5, size=1):
    """
    binomial distribution is the discrete probability distribution of a specific number of successes in a sequence of n independent experiments, 
    each asking a yes-no question, and each with its own boolean-valued outcome: 
    success with probability p or failure with probability q = 1 − p.
    """
    
    b = []
    for _ in range(size):
        u = uniform(size=n, seed=generate_seed()) # Binomial random variate generator from the Uniform generator
        y = (u<=p).astype(int)
        b.append(np.sum(y))
    return b
```

```python
ls = binomial_distribution(p=0.6,size=10000)
plt.hist(ls, bins=20, edgecolor='k')
plt.show()
```

### Poission distribution

```python
def poisson_distribution(alpha, size=1):
    """
    poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring 
    in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event.
    """
    
    poisson = []
    for k in range(size):
        u = uniform(size=5*alpha, seed=generate_seed()) # poission random variate generator from the Uniform generator
        x,p,i = 0,1,0
        while p>=np.exp(-alpha):
            p = u[i]*p
            x+=1
            i+=1
        poisson.append(x)
    return poisson
```

```python
ls = poisson_distribution(20, 10000)
plt.hist(ls, bins=20, edgecolor='k')
plt.show()
```

### Exponential distribution

```python
def expo_distribution(lambda_val, size=1):
    u = uniform(size=size, seed=generate_seed())
    return -(1/lambda_val)*(np.log(1-u))
```

```python
ls = expo_distribution(0.6, 10000)
plt.hist(ls, bins=20, edgecolor='k')
plt.show()
```

## Predicting the value of pi using generated number

```python
def pi_estimator(samples):
    points_inside_circle= 0
    total_num_points = 0
    X,Y = pseudo_uniform(size=2*samples).reshape(2,-1)

    for x,y in zip(X,Y):
        distance = x**2 + y**2
        if distance <= 1:
            points_inside_circle +=1
        total_num_points += 1
    return 4*points_inside_circle/total_num_points
```

```python
pi_pred = pi_estimator(10**7)
print('Predicted value of pi is', pi_pred)
```

```python
pi_test = np.pi
error = abs(pi_test-pi_pred)/pi_test
print(error)
```
